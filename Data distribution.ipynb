```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('prepared_bank_data.csv')
```

```python
numeric_cols = ['age', 'balance', 'day', 'duration', 'campaign', 'pdays', 'previous']
for col in numeric_cols:
    plt.figure(figsize=(8,4))
    sns.histplot(df[col], bins=30, kde=True)
    plt.title(f'{col}distribution')
    plt.xlabel(col)
    plt.ylabel('frequency')
    plt.show()
```

```python
categorical_vars = [
    'default', 'housing', 'loan', 'marital_married', 'marital_single', 'marital_divorced',
    'education_primary', 'education_secondary', 'education_tertiary',
    'contact_cellular', 'contact_telephone', 'contact_unknown',
    'poutcome_failure', 'poutcome_other', 'poutcome_success', 'poutcome_unknown',
    'season', 'season_autumn', 'season_spring', 'season_summer', 'season_winter'
]

for col in categorical_vars:
    plt.figure(figsize=(6,4))
    counts = df[col].value_counts()
    counts.plot(kind='bar')
    plt.title(f'{col} Category distribution (bar chart)')
    plt.xlabel(col)
    plt.ylabel('frequency')
    plt.show()

    plt.figure(figsize=(6,6))
    counts.plot(kind='pie', autopct='%1.1f%%')
    plt.title(f'{col} Category proportions (pie chart)')
    plt.ylabel('')
    plt.show()
```
